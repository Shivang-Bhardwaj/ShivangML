```python
df['LotConfig'].unique() #Series.unique()-->return array of unique elements.
```

```python
df[df['LotConfig']=='Inside']['LotFrontage']
#agar hamein kisi bhi numeric variable ki missing values ko impute karna ha tto hamein kisi ek categorical variable ko choose 
 # karna hoga.konsa categorical variable choose karna ha iiske liye hamein domain knowledge hona jaroori ha.
    

#df[df['LotConfig']=='Inside']-->'LotConfig' feature mein jaha-jaha 'Inside' as a datavalue hoga, vahi rows milengi.
#df[LotConfig']=='Inside'-->series ke har ek data value par (==)relational operator lagega.
  #yaha ek series dega of True & False as data values.aap consider kar sakte ha ki hamari series [] mein band ha.
```

```python
df[df['LotConfig']=='Inside']['LotFrontage'].replace(np.nan,df[df['LotConfig']=='Inside']['LotFrontage'].mean())
```

```python
df_copy=df.copy()# df.copy() is used becoz agar use nahi karta tto df_copy df ki image ban jati.
for var_class in df['LotConfig'].unique():
        df_copy.update(df[df['LotConfig']==var_class]['LotFrontage'].replace(np.nan,df[df['LotConfig']==var_class]['LotFrontage'].mean()))
    #or  df.update(df[df[LotConfig']==var_class]['LotFrontage'].replace(np.nan,df[df['LotConfig']==var_class]['LotFrontage'].mean()))
        #agar hum ussi dataframe ko update karna chahate ha.
    #hum np.nan ko kisi 'Global Constant' ke saath bhi replace kara sakte ha.
        
        
# df.update()-->to update dataframe
```

```python
df_copy.isnull().sum()
```

```python
df_copy=df.copy()
num_vars_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig','MasVnrType','GarageType']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df[cat_var]==var_class][num_var_miss].replace(np.nan,df[df[cat_var]==var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

```python
df_copy[df_copy['MasVnrType'].isnull()==True] #checking kya 'MasVnrType' mein nan values ha. jis categorical variable ke
# according hum value impute kar rahe ha,agar vahi dirty ha tto value impute nahi ho paegi.uss variable ka clean hona jaroori ha
```

```python
df_copy[df_copy['GarageType'].isnull()==True]
```

```python
df_copy=df.copy()
num_vars_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df[cat_var]==var_class][num_var_miss].replace(np.nan,df[df[cat_var]==var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

```python
df_copy_median=df.copy()
num_vars_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df[cat_var]==var_class][num_var_miss].replace(np.nan,df[df[cat_var]==var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
#--------------------------------data distribution->for checking accuracy of cleaned dataset------------------------------------
```

```python
plt.figure(figsize=(10,10))
sns.set()
for index,var in enumerate(num_vars_miss):
    plt.subplot(2,2,index+1)
    sns.distplot(df[var],bins=20,kde_kws={'linewidth':8,'color':'r'},label='original')
    sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':5,'color':'g'},label='cleaned')
    plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for index,var in enumerate(num_vars_miss):
    plt.subplot(2,2,index+1)
    sns.distplot(df[var],bins=20,kde_kws={'linewidth':8,'color':'r'},label='original')
    sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':5,'color':'g'},label='mean')
    sns.distplot(df_copy_median[var],bins=20,kde_kws={'linewidth':3,'color':'k'},label='median')
    plt.legend()
```

```python
for i, var in enumerate(num_vars_miss):# to check outliers is increased or decreased.
    plt.figure(figsize=(10,10))
    plt.subplot(311)        #outliers-->value that 'lies outside'(is much smaller or larger than) other datavalues in dataset. 
    sns.boxplot(df[var])    #eg->score:25,29,3,32,85,33,27,28.  both 3 and 85 are outliers.
    plt.subplot(312)
    sns.boxplot(df_copy[var])
    plt.subplot(313)
    sns.boxplot(df_copy_median[var])
```
