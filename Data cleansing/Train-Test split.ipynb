```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df2=df[['survived','pclass','age','parch']]
df3=df2.fillna(df2.mean())
df3.head()
```

```python
X=df3.drop(columns='survived')
y=df3['survived']
print('Shape of X=',X.shape) #dependent (target variable) and independent variables ko alag karna hi padta ha.
print('Shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=58)#-->split array or dataset into train & test subsets.
print('Shape of X_train=',X_train.shape)               #train_size= 
print('Shape of y_train=',y_train.shape)
print('Shape of X_test=',X_test.shape)
print('Shape of y_test=',y_test.shape)

#test_size=0.2-->0.2*100(or 20%)=>kitne % dataset as test dataset chahiye. rest 80% dataset as a train dataset chahiye.
#if you don't mention random_state in code then whenever you execute your code,your train & test subsets will get different values each time.
```
