```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['Python','R','AI','ML','DS']
c1_stu=[23,9,20,25,10]   #list, tuple, array   #out of 100 student in each class
c2_stu=[40,5,20,20,10]  
c3_stu=[35,5,30,15,15]
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.3
plt.barh(classes_index,c1_stu,width,color='c',edgecolor='b',linestyle='--',align='edge',linewidth=2,alpha=0.5)
plt.barh(classes_index+width,c2_stu,width,color='m',edgecolor='g',linestyle='--',align='edge',linewidth=2,alpha=0.5)
plt.barh(classes_index+width*2,c3_stu,width,color='y',edgecolor='r',linestyle='--',align='edge',linewidth=2,alpha=0.5)
plt.yticks(classes_index,classes,rotation=180)
plt.legend(['class1 stu','class2 stu','class3 stu'])
plt.title('Class student data',fontsize=15)
plt.ylabel('classes',fontsize=13)
plt.xlabel('No. of student',fontsize=13)
plt.show()
```
