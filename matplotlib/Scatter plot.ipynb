```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
plt.figure(figsize=(16,9))
plt.scatter(tips_df.day,tips_df.total_bill)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(tips_df.tip,tips_df.total_bill)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(tips_df.tip,tips_df.total_bill,marker='*',label='tips data')
plt.title('Scatter plot',fontsize=16)
plt.xlabel('Tips',fontsize=13)
plt.ylabel('Total Bill',fontsize=13)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(tips_df.tip,tips_df.total_bill,c='r',s=100,linewidths=2,edgecolors='y',alpha=1)
plt.title('Scatter plot',fontsize=16)
plt.xlabel('Tips',fontsize=13)
plt.ylabel('Total Bill',fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(tips_df.tip,tips_df.total_bill,c='w',s=[100,500,900],linewidths=2,edgecolors=['y','b','r','g'],alpha=1)
plt.title('Scatter plot',fontsize=16)
plt.xlabel('Tips',fontsize=13)
plt.ylabel('Total Bill',fontsize=13)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(16,9))
arr=np.random.randint(1,50,244)
plt.scatter(tips_df.tip,tips_df.total_bill,c='r',s=200,linewidths=2,edgecolors='c',alpha=1,label='Bill')
plt.scatter(tips_df.tip,arr,c='y',s=50,linewidths=1,edgecolors='k',marker='d',alpha=1,label='Practical')
plt.title('Scatter plot',fontsize=16)
plt.legend(loc=4)
plt.show()
```
