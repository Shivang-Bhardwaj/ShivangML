```python
#scatter plot
import matplotlib.pyplot as plt
import pandas as pd
```

```python
classes=['py','R','ML','AI','DS']
c1_stu=[45,15,35,25,30]
plt.pie([45]) #to plot single info.single info hamesha circular format mein dikhegi
plt.show()
```

```python
classes=['py','R','ML','AI','DS']
c1_stu=[45,15,35,25,30]
plt.pie(c1_stu,labels=classes)#labels ke bina hamm sirf colors ko dekhkar info ko samjh nahi sakte  
plt.show()
```

```python
classes=['py','R','ML','AI','DS']
c1_stu=[45,15,35,25,30]
x=[0.1,0,0.25,0,0] #untne 0 dene ha jitne classses mein data ha. yaha no. distance from other data ko represent karta ha.
colors=['c','b','r','y','g']# jitne pieces utne colors.hum apni marji se bhi colors de sakte ha
plt.pie(c1_stu,labels=classes,explode=x,colors=colors,autopct='%0.1f%%',shadow=True,radius=1.4)
#radius=to change size of circle
#shadow=False(by default)
# %0.1f%%-->f=float,0.1=point ke baad 1 digit,agar 0.2 likhege tto point ke baad 2 digit tak no. ...etc
#explode hum kisi data ko slice karne mein use karte ha
plt.show()
```

```python
plt.figure(figsize=(4,4))
wedgeprops={'linewidth':2.2,'width':0.8,'edgecolor':'k'}
textprops={'fontsize':15}#jo bhi text likha ha usmein changes lana
plt.pie(c1_stu,labels=classes,autopct='%0.1f%%',shadow=True,startangle=270,textprops=textprops,pctdistance=0.8,
    labeldistance=1.4,counterclock=True,wedgeprops=wedgeprops,frame=True,rotatelabels=True,center=(1,1))
#bydefault 1st part start from 0(deg) but we can start it from any angle(in anticlockwise direction).
#pctdistance(or %distance)=likhe huae % ki distance center se(value in float format)
#labeldistace=label ki distance from center(in float format)
#counterclock=plotting of graph in which direction(by default True)
#hame pahale frame=True karna hoga tabhi hum center parameter ka use kar payege to locate center
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,4))
colors=['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels=np.ones(20)
plt.pie([1],colors='k',radius=2.05)
plt.pie(labels,colors=colors,radius=2.0)

plt.pie([1],colors='g',radius=1.8)
plt.pie([1],colors='y',radius=1.6)
plt.pie([1],colors='c',radius=1.3)
plt.pie([1],colors='b',radius=1.1)
plt.pie([1],colors='m',radius=0.9)

plt.pie([1],colors='b',radius=0.31)
plt.pie(labels,colors=colors,radius=0.3)

plt.pie([1],colors='w',radius=0.2)
plt.pie([1],colors='k',radius=0.1)
plt.show()
```
