```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ml=np.random.randint(18,40,100)
py=np.random.randint(19,50,100)
print(ml)
print(py)
```

```python
from matplotlib import style
style.use('ggplot')
bins=[15,20,25,30,35,40,45,50]
plt.hist([ml,py],bins,rwidth=0.8,color=['c','m'],orientation='horizontal',label=['ML Student','Py Student'],alpha=0.5)
plt.title('ML Students',fontsize=15)
plt.xlabel('No. of Student',fontsize=13)
plt.ylabel('Student Age Group',fontsize=13)
plt.grid(color='y',linestyle=':',linewidth=2)
plt.legend()
plt.show()
```
