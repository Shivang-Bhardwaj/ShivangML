```python
from sklearn.linear_model import Ridge,Lasso
import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df=pd.read_csv(r'C:\Users\DELL PC\Desktop\ML\Useful data\Bengaluru_House_Data.csv',nrows=1000)
df.head()
```

```python
df=df.dropna()
series=df['availability'].value_counts()
series
```

```python
for i in series.keys():
    if i!='Ready To Move':
        df['availability'].replace(i,'no',inplace=True)
```

```python
df1=pd.get_dummies(df,drop_first=True)
```

```python
X=df1.drop(columns='price')
y=df1['price']
print('Shape of X=',X.shape)
print('Shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=58)
print('Shape of X_train=',X_train.shape)
print('Shape of y_train=',y_train.shape)
print('Shape of X_test=',X_test.shape)
print('Shape of y_test=',y_test.shape)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
rd=Ridge()#alpha=1.0
rd.fit(X_train,y_train)
```

```python
rd.coef_
```

```python
rd.intercept_
```

```python
rd.predict(X_test)
```

```python
y_test
```

```python
rd.score(X_test,y_test)
```

```python
#-------------------------------------------------------------------------------------------------------------------------------
ls=Lasso()#alpha=1.0(by default)
ls.fit(X_train,y_train)
```

```python
ls.coef_
```

```python
ls.intercept_
```

```python
ls.predict(X_test)
```

```python
y_test
```

```python
ls.score(X_test,y_test)
```

```python
ls1=Lasso(alpha=2)
ls1.fit(X_train,y_train)
ls1.score(X_test,y_test)
```

```python
#Lasso>Ridge>linear
#Lasso %accuracy increase kar deta ha.alpha ki value ko increase karne se bhi accuracy par fark padata ha
```

```python
y_pred=ls.predict(X_test)
```

```python
#-------------------------------------------which algo. to choose --------------------------------------------------------------
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print('MSE=',mse)
print('RMSE=',rmse)
```

```python
#----------------------------------------R-Squared------------------------------------------------------------------------------
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
