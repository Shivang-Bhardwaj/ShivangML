```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df=pd.read_csv(r'C:\Users\DELL PC\Desktop\ML\Useful data\Bengaluru_House_Data.csv',nrows=1000)
df.head()
```

```python
df.select_dtypes(include=['object']).columns
```

```python
df=df.dropna()
series=df['availability'].value_counts()
series
```

```python
for i in series.keys():
    if i!='Ready To Move':
        df['availability'].replace(i,'no',inplace=True)
```

```python
df1=pd.get_dummies(df,drop_first=True)
```

```python
df1.shape
```

```python
df1.head()
```

```python
X=df1.drop(columns='price')
y=df1['price']
print('Shape of X=',X.shape)
print('Shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=58)
print('Shape of X_train=',X_train.shape)
print('Shape of y_train=',y_train.shape)
print('Shape of X_test=',X_test.shape)
print('Shape of y_test=',y_test.shape)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
#ml model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()#-->this class for both linear & multiple regression

lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#predict the value of Home & Test
X_test[0,:]
```

```python
lr.predict([X_test[0,:]])
```

```python
lr.predict(X_test)#-->to predict value of target variable.
```

```python
y_test# to check predicted values.
```

```python
lr.score(X_test,y_test)# X_test ki madad se value predict kar aur y_test se compare karke %Accuracy find kar.
# Ans->0.79678 means 79.678% accurate ha hamara model
```
