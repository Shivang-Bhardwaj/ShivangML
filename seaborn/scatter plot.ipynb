```python
import seaborn as sns,matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df
```

```python
sns.scatterplot(data=titanic_df,x='age',y='fare')
# achi tarah se visualise nahi kar pa rahe ha 
```

```python
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex')
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='age',y='fare',style='who') 
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex',style='who') 
#hue,style,size->category ke according scatter plot karne mein madad karte ha
# hue ->produce points with different color
#style->produce points with different markers
#size-> produce poits with different sizes
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex',style='who',size='who') 
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex',style='who',size='who',sizes=(100,400)) 
#sizes=(min value,max value)->to set sizes to markers.it is used when size parameter is used.agar size parameter use nahi kiya
# ha tto 'sizes' apna kuch bhi effect nahi dikhaega. see info
```

```python
#to find  who is alive 
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='who',y='fare',hue='alive',style='alive',size='who',sizes=(100,400))
#here x=categorical data

#info-->max woman & child are saved but 95%man are not saved.women & children who pay less fare are not alive
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=titanic_df,x='who',y='fare',hue='alive',style='alive',size='who',sizes=(100,400),palette='magma')
```

```python
#just for fun->not for machine learning point of view
plt.figure(figsize=(16,9))
sns.scatterplot(x='age',y='fare',data=titanic_df)
sns.lineplot(x='age',y='fare',data=titanic_df)
sns.barplot(x='age',y='fare',data=titanic_df)
```

```python
plt.figure(figsize=(16,9))
kwargs={'linewidth':1,'edgecolor':'k','color':'r','linestyle':'--','alpha':1}
sns.scatterplot(data=titanic_df,x='sex',y='fare',size='sex',sizes=(100,400),**kwargs)
```
