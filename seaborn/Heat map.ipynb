```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt,seaborn as sns
```

```python
arr_2d=np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)#take only 2-d dataset(2d dataset->containing rows and columns).only work on numeric value in color box format
#2d dataset ke har ek data(cell) ko color se represent karega.color of colorbox show corresponding value of dataset using colorbar
#indexes(or labels) of rows & columns of 2d dataset will be used as labels for y&x axis of heat map respectively.  
```

```python
globalwarming_df=pd.read_csv(r'C:\Users\DELL PC\Desktop\ML\Useful data\globalwarming.csv')
globalwarming_df
#yaha numeric values show kar rahi ha ki kitni co2, ek country release karti ha. more value->more co2 release
```

```python
globalwarming_df=globalwarming_df.dropna()#dropping rows containing NaN values
globalwarming_df=globalwarming_df.drop(columns=['Country Code','Indicator Name','Indicator Code']).set_index('Country Name')
globalwarming_df.head(6)
#df.set_index()->kis column ki values ko as a row index banana ha
#we drop 'Country Code','Indicator Name','Indicator Code' becoz heat map can only represent numeric data.
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df.head(6))

#Info-->Albania bahut kam co2 release karta ha.Aruba ne apni halat sudhari->co2 release kam karta gaya
#      Afghanistan tto co2 release badhata ja raha ha
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df.head(6),vmin=0,vmax=30)
#vmin,vmax->to fix lower and upper value of colorbar
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df.head(6),cmap='coolwarm')
#cmap->to change color of heat map."coolwarm" sabse best 'cmap value' hoti ha for plotting heatmap.  cool->blue,warm->red
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df.head(6),annot=True)#annot->colorbox ke corresponding dataset ki value ko dikhana ha ya nahi
#colorbox mein fill ki gayi values ko 'annot data'  khate ha
```

```python
annot_arr=np.array([['a00','a01','a02'],['a10','a11','a12'],['a20','a21','a22'],['a30','a31','a32']])
annot_arr
```

```python
sns.heatmap(arr_2d,annot=annot_arr,fmt='s')
#If True, write the data value in each cell.If an array 'arr' with the same shape as ``data``is written in 'annot',then original
#data values ki jagah 'arr' mein likha hui data values color box mein aa jaegi.but we have to specify ki 'arr' mein likha hua
#data kis format mein ha->to specify,use "fmt"(format)='s'/'f'/'d'.    's'->string,'f'->float,'d'->int
```

```python
sns.heatmap(arr_2d,annot=annot_arr,fmt='s',square=True)
#square=False(by default)-->set shape of each cell of heatmap in square
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':10,
           'fontstyle':'italic',   #-->normal,italic
          'color':'k',             #-->for color of text
           'backgroundcolor':'w',
          'rotation':'vertical',    #-->horizontal,vertical
          'verticalalignment':'center',   #'top', 'bottom', 'center', 'baseline', 'center_baseline'
          'alpha':0.7}   #for transparency
sns.heatmap(globalwarming_df.head(6),annot=True,annot_kws=annot_kws)
# annot_kws-->to dacorate annot data
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':10,
           'fontstyle':'italic',
          'color':'k',            
           'backgroundcolor':'w',
          'rotation':'horizontal',   
          'verticalalignment':'top',
          'alpha':0.7}   
sns.heatmap(globalwarming_df.head(6),annot=True,annot_kws=annot_kws,cmap='coolwarm',linewidths=4,linecolor='k')
#linewidths=0-->sabhi colorboxes ke beech line create karna.yaha colorboxes ko ek dusare se alag karne mein madad karta ha.
#linecolor='w'(by default)-->to fill color in line
```

```python
sns.heatmap(globalwarming_df.head(6),cbar=False,xticklabels=False,yticklabels=False)
#cbar=True(by default)-->color bar chahiye ya nahi
#xticklabels=True(by default),yticklabels=True(by default)-->x & y lables ko hatana ha ya nahi.

#'Country Name' rah gaya becoz yaha indexes ke column ka naam tha and agar isse hatana ha tto dataframa mein changes karne honge
```

```python
plt.figure(figsize=(16,9))
country_code=['ABW','AFG','AGO','ALB','AND','ARB']
sns.heatmap(globalwarming_df.head(6),xticklabels=np.arange(0,15),yticklabels=country_code)
#xticklabels=True/False/a.  'a'->list,tuple,array. 'a' ke andar vo items(str or no.) ha jinhe x-axis par likhe labels ke saath replace karna ha.
     #condition->no. of items in 'a' must be equal to no. of labels on x-axis
#yticklabels->same as "xticklabels"
```

```python
plt.figure(figsize=(16,9))
cbar_kws={'orientation':'horizontal', #horizontal,vertical(by default)
         'shrink':1,  #for colorbar size
         'extend':'min', #extend='min'/'max'/'both'  -->birthday vali topi create karne ke liye
         'extendfrac':0.1,   #birthday vali topi ki length ke liye.agar ham "extend" parameter ka use nahi karenge tto yaha parameter kaam nahi karega 
         'ticks':np.arange(0,26),  #colorbar ki labelling ke liye
         'drawedges':True}   #to draw lines on colorbar
sns.heatmap(globalwarming_df.head(6),cbar_kws=cbar_kws)
```

```python
#         multiple heatmaps using "plt.subplot()" 
plt.figure(figsize=(30,10))
plt.subplot(131)
sns.heatmap(globalwarming_df.head(6),cbar=False,linecolor='k',linewidths=1)

plt.subplot(132)
sns.heatmap(globalwarming_df.head(6),cbar=False,linecolor='c',linewidths=1)

plt.subplot(133)
sns.heatmap(globalwarming_df.head(6),cbar=False,linecolor='y',linewidths=1)
plt.show()
```

# -------------------------------------Correlation Heatmap-------------------------------------------
### used in feature engineering & feature scaling

```python
globalwarming_df.corr() #--->give a dataframe which is called correlation matrix
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(globalwarming_df.corr(),annot=True,linewidths=3)
ax.tick_params(size=10,color='c',labelsize=10,labelcolor='c')# params->parameters
plt.title('Heat map of "Who is responsible for global warming"',fontsize=25)
plt.show()


#info->khudh ka khudh ke saath strong co-relation hota ha.so,diagonal ke sabhi elements '1' ha.diagonal ke oopar and neche ka data
#mirror image hota ha,tto ya tto diagonal ke oopar vala data ka ya neche vala data ka analysis karte ha.

#har ek year ka correlation 2014 ke saath aur strong hota ja raha ha=>
```

```python
plt.figure(figsize=(16,9))
corr_mx=globalwarming_df.corr()
sns.heatmap(corr_mx,mask=True)
#mask=True/False/a.  a->array,dataframe le sakta ha. mask-->mask mein jo bhi values dalenge(in form of array or dataframe) tto
#uske corresponding colorbox nahi dikhai denga.'a' mein vo annot values daalni ha jo "data" mein ho.
```

```python
#upper triangle heatmap
plt.figure(figsize=(16,9))
corr_mx=globalwarming_df.corr()  #correlation matrix
matrix=np.tril(corr_mx) #take lower correlation matrix
sns.heatmap(corr_mx,mask=matrix)
```

```python
#lower triangle heatmap
plt.figure(figsize=(16,9))
corr_mx=globalwarming_df.corr()  #correlation matrix
matrix=np.triu(corr_mx) #take upper correlation matrix
sns.heatmap(corr_mx,mask=matrix)
```

```python
from sklearn.datasets import load_breast_cancer #sklearn->'scikit learn'->a ML library,a collection of 'ml' algorithms
cancer_dataset=load_breast_cancer()# sklearn->package  datasets->module
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                       columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidths=3)
ax.tick_params(size=10,color='c',labelsize=10,labelcolor='c')
plt.title('Heat map of "Breast Cancer Patient"',fontsize=25)
plt.show()
```
