```python
#histogram/distplot
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['size'])#see info
#histogram ke saath curve bhi mila ha
#curve represent KDE->kernel density estimation
```

```python
sns.distplot(tips_df['total_bill'],color='r')
```

```python
sns.distplot(tips_df['total_bill'],hist=False)#hist=true(by default) histogram rakhna ha ki nahi,to get only KDE
```

```python
sns.set(style='white')
sns.distplot(tips_df['total_bill'],kde=False)# kde=True(by default) curve hatana ha ki nahi
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],kde=False,bins=10)
```

```python
sns.distplot(tips_df['total_bill'],rug=True,kde=False,fit=norm)
#rug=false(by default) to plot small bar.these bars represent values of total_bill
#norm->normalised line plot.norm ko use karne ke liye scipy library se norm ko import karana hoga
```

```python
sns.distplot(tips_df['total_bill'],vertical=True)
#vertical=false(by default)
```

```python
sns.set()
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df['total_bill'],axlabel='Total Bill',label='bill')
plt.legend()
plt.title('histogram')
#axlabel=x ya y axis ko naam dena ha tto
#label ko use karna ha tto plt.legend likhna jaroori ha
```

```python
tips_df.total_bill.sort_values()#sort_values()->func
```

```python
sns.set()
plt.figure(figsize=(16,9))
bins=[1,5,10,15,20,25,30,35,40,45,50,55]#total_bill ki sorting kar max and min value dekhkar hamne apne hisaab se groups banaae
#ha.
sns.distplot(tips_df['total_bill'],bins=bins)
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],bins=bins)
plt.xticks(bins)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],hist_kws={'color':'r','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':1},
            kde_kws={'color':'g','linewidth':5,'linestyle':':','alpha':1},
            rug=True,
            rug_kws={'color':'c','edgecolor':'m','linewidth':1,'linestyle':'--','alpha':1})
#hist_kws=to decorate bars.  kws->keyword arguments
#kde_kws=to decorate kde curve
#rug_kws isko use karne ke liye rug =true karna hoga
```

```python
sns.set(style='dark')
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],kde=False,fit=norm,
             fit_kws={'color':'r','linewidth':5,'linestyle':'--','alpha':1})
#fit_kws ka use karna ha tto 'fit' parameter ka use karna hoga.
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],bins=9,label='Total Bill',kde=False)
sns.distplot(tips_df['tip'],bins=9,label='Tip',kde=False)
sns.distplot(tips_df['size'],bins=9,label='Size',kde=False)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
g=sns.distplot(tips_df['total_bill'],bins=9,label='Total Bill',kde=False)
g.set(title='B_bill',xlabel='Size',ylabel='Probability Density')
plt.show()
```
